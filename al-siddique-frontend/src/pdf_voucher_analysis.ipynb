# PDF Voucher Template Analysis and Multi-Template Generator

This notebook analyzes the provided fee voucher PDF, extracts layout and image details, and builds three voucher templates: one exact replica and two modern designs. The final section includes a selectable settings mechanism so the user can choose between the templates.

## 1. Analyze PDF Structure

Load the PDF and inspect pages, dimensions, and block structure for the voucher template.

```python
import os
from pathlib import Path
pdf_path = Path(r'C:/Users/Mac&lap/Downloads/Print Students Vouchers12.pdf')
print('PDF path exists:', pdf_path.exists())
print('PDF absolute path:', pdf_path.resolve())

import fitz

doc = fitz.open(pdf_path)
print('Number of pages:', len(doc))
for page_number, page in enumerate(doc, start=1):
    print(f'--- Page {page_number} ---')
    print('Page rect:', page.rect)
    text_dict = page.get_text('dict')
    print('Blocks count:', len(text_dict['blocks']))
    if len(text_dict['blocks']) > 0:
        first_block = text_dict['blocks'][0]
        print('First block type:', first_block['type'])
        if 'lines' in first_block:
            first_line = first_block['lines'][0]
            if first_line['spans']:
                first_span = first_line['spans'][0]
                print('First span text:', first_span['text'])
                print('Font size:', first_span['size'])
                print('Font flags:', first_span['flags'])
    print()
```

## 2. Extract Text and Layout Elements

Extract text positions, font metadata, and table-like structure from the voucher page.

```python
page = doc[0]
text_blocks = page.get_text('blocks')
print('Total text blocks:', len(text_blocks))
for i, block in enumerate(text_blocks[:20], start=1):
    x0, y0, x1, y1, text, block_type = block
    print(f'Block {i}: type={block_type}, rect=({x0:.1f}, {y0:.1f}, {x1:.1f}, {y1:.1f})')
    print('  text sample:', repr(text.strip().replace('\n', ' ')[:120]))
    print()

images = page.get_images(full=True)
print('Images count:', len(images))
if images:
    for img_index, img in enumerate(images, start=1):
        print('Image', img_index, img)
```
